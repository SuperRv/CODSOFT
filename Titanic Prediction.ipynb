## Handling missing values

```python
#drop cabin column as most of the values are missing
data = data.drop('Cabin', axis=1)
data
```

```python
data.isnull().sum()
```

```python
#Using the same method to add missing values in age column
data.groupby(['Pclass', 'Sex']).mean()
```

```python
data['Age'] = round(data['Age'].fillna(data.groupby(['Pclass', 'Sex'])['Age'].transform('mean')),2)
```

```python
data.isnull().sum()
```

```python
data.groupby('Embarked').count()
```

```python
# Since more than 50% of passangers have embarked from 'S' we will fill 2 missing values with 'S'
data['Embarked'].fillna('S', inplace=True)
```

## Model Training and Building

We have 2 coulmn which can be converted to numerical values i.e. Sex and Embarked

Sex - Female-0, Male-1<br>
Embarked - C-0, Q-1, S-2

```python
#Using Label Encoding for the same
from sklearn.preprocessing import LabelEncoder
```

```python
data['Sex'].head()
```

```python
data['Sex'] = LabelEncoder().fit_transform(data['Sex'])
```

```python
data['Sex'].head()
```

```python
data['Embarked']
```

```python
data['Embarked'] = LabelEncoder().fit_transform(data['Embarked'])
```

```python
data['Embarked']
```

```python
data
```

```python
#Splitting the data into training and testing set
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(columns=['PassengerId', 'Survived', 'Name', 'Ticket'], axis =1)
Y = data['Survived']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
lr_model = LogisticRegression().fit(X_train,Y_train)
lr_pred = lr_model.predict(X_test)
con = classification_report(Y_test, lr_pred)
print(con)
```
